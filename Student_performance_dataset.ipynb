```python
df=df.drop(['school'],axis=1)
```

```python
df=df.drop(['address'],axis=1)
```

```python
df=df.drop(['failures'],axis=1)
```

```python
import pandas as pd
import numpy as np


# Define features (x) and target (y)
# For simplicity, let's start with only numeric features.
# You might want to encode categorical features later for a more comprehensive model.
numeric_df = df.select_dtypes(include=[np.number])
x = numeric_df.drop('g3', axis=1)
y = df['g3']

```

```python
x.columns
```

```python
y.name
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
y_train.value_counts(normalize=True)
```

```python
df.value_counts(normalize=True)
```

```python
df['g3'].value_counts(normalize=True)
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train[:5]
```

```python
# implementing model
# linear Regression
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# prediction
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
y_pred_test_dt=model.predict(x_test)
y_pred_train_dt=model.predict(x_train)
```

```python
# Accuracy
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred_test_dt))
print(r2_score(y_train,y_pred_train_dt))
```

```python
# implementing another model
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier(random_state=42)
model2.fit(x_train,y_train)
```

```python
# prediction
y_pred_test_dt=model.predict(x_test)
y_pred_train_dt=model.predict(x_train)
```

```python
# evaluation accuracy
from sklearn.metrics import r2_score

print(r2_score(y_train,y_pred_train_dt))
print(r2_score(y_test,y_pred_test_dt))
```

```python
# imp0lementing a new model randomforestclassifier
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier(random_state=42,n_estimators=100)
model3.fit(x_train,y_train)
```

```python
# prediction
y_pred_test_rf=model.predict(x_test)
y_pred_train_rf=model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score

# Correctly generate predictions from the RandomForestClassifier (model3)
y_pred_test_rf = model3.predict(x_test)
y_pred_train_rf = model3.predict(x_train)

print(accuracy_score(y_test, y_pred_test_rf))
print(accuracy_score(y_train, y_pred_train_rf))
```

```python
# confusion matrix
print(confusion_matrix(y_test,y_pred_test_rf))
```

```python
# classification report
print(classification_report(y_test,y_pred_test_rf))
print(classification_report(y_train,y_pred_train_rf))
```
